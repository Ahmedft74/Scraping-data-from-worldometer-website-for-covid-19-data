```python
import requests as req 
from bs4 import BeautifulSoup as bs
import numpy as np 
import pandas as pd 
```

```python
html = req.get('https://www.worldometers.info/coronavirus/')
```

```python
html_parsed = bs(html.content)
```

```python
table = html_parsed.find('table', attrs = {'id': 'main_table_countries_today'}) 
```

```python
table
```

```python
rows = table.find_all('tr')
```

```python
rows[0].text.strip().split('\n')
```

```python
data = []

for x in rows:
    data.append(x.text.strip().split('\n')[1:9])

df = pd.DataFrame(data)

df.head()
```

```python
df=pd.DataFrame(data[9:],columns=data[0])
```

```python
df.rename(columns={'Country,Other': 'Country', 'TotalCases': 'Total Cases','NewCases': 'New Cases','TotalDeaths': 'Total Deaths','NewDeaths': 'New Deaths','TotalRecovered': 'Total Recovered','NewRecovered': 'New Recovered','ActiveCases' : 'Active Cases'}, inplace=True)
```

```python
df = df.set_index('Country')
```

```python
df.to_csv('covid19.csv')
```
